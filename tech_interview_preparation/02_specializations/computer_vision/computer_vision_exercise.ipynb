## Ejercicio 1: Cargar CIFAR-10

```python
# Cargar dataset
(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()

print(f"Train shape: {X_train.shape}")
print(f"Test shape: {X_test.shape}")

# Normalizar
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Flatten labels
y_train = y_train.flatten()
y_test = y_test.flatten()

# Visualizar
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[i])
    ax.set_title(classes[y_train[i]])
    ax.axis('off')
plt.tight_layout()
plt.show()
```

## Ejercicio 2: Fully Connected vs CNN

```python
# Fully Connected (Baseline)
model_fc = keras.Sequential([
    layers.Flatten(input_shape=(32, 32, 3)),
    layers.Dense(512, activation='relu'),
    layers.Dense(256, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# CNN (Mejor)
model_cnn = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

print("Fully Connected - Parameters:")
print(f"  Total: {model_fc.count_params():,}")
print("\nCNN - Parameters:")
print(f"  Total: {model_cnn.count_params():,}")
print(f"\n💡 CNN tiene {model_fc.count_params() / model_cnn.count_params():.1f}x menos parámetros")
```
